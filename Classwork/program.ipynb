```python
import math

# take radius from user
r = float(input("Enter radius: "))

# calculate area and perimeter
area = math.pi * r * r
perimeter = 2 * math.pi * r

# display result
print("Area of circle =", area)
print("Perimeter of circle =", perimeter)
```
