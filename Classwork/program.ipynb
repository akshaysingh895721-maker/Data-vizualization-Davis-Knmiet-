```python
num = int(input("Enter a number: "))
if num == 0:
    print("0 has infinite factors")
elif num < 0:
    print("Negative number find")
    num = abs(num)

# Create array from 1 to num
arr = np.arange(1 ,num+1)

#find factors
factors = arr[num % arr == 0]

print("All factors:", factors)
#concept for even factor
even_factors = factors[factors % 2 == 0]
print("Even factor",even_factors)

```
