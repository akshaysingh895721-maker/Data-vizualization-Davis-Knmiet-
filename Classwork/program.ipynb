```python
# Open a file in write mode
f = open("output.txt", "w")

# Create an empty list to store sentences
lines = []

# Use range() to generate 20 sentences
for i in range(1, 21):
    # Add each sentence with a newline character
    lines.append("Sentence " + str(i) + "\n")

# Write all sentences to the file at once
f.writelines(lines)

# Close the file
f.close()
```
