```python
#Program to calculate Simple Interest

#Taking the input from the user
p = float(input("Enter Principal: "))
r = float(input("Enter Rate: "))
t = float(input("Enter Time: "))

si = lambda p, r, t: (p * r * t) / 100
print("Simple Interest:", si(p, r, t))
```
