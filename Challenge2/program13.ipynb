```python
#program to Count Numbers Greater Than 50


#take the input from the user
nums = list(map(int, input("Enter numbers separated by space: ").split()))

#apply the condition of the program
count = 0
for num in nums:
    if num > 50:
        count += 1

print("Count =", count)
```
