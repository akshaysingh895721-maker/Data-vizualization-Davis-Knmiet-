```python
#program to find the duplicate word in the string

#take the input from the user
text = input("Enter a string: ").lower()
seen = set()
duplicates = set()

for ch in text:
    if ch in seen:
        duplicates.add(ch)
    else:
        seen.add(ch)

print("Duplicate characters:", " ".join(sorted(duplicates)))
```
