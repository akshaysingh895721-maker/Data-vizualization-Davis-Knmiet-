```python
#program to count the occurence of the character

#take the input from the user
text = input("Enter a string: ")
char = input("Enter character to count: ")
count = 0

#logic of the count the occurence of character
for ch in text:
    if ch == char:
        count += 1

print("Count =", count)
```
