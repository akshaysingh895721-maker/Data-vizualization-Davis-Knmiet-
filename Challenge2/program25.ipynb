```python
#program to convert the string in uppercase

#take the input from the user
text = input("Enter a string: ")
result = ""

for ch in text:
    # Check if lowercase letter
    if 'a' <= ch <= 'z':
        # Convert to uppercase by subtracting 32 from ASCII
        result += chr(ord(ch) - 32)
    else:
        result += ch

print("Output:", result)
```
