```python
#program to identify grade bassed on mark

#take the input from the user
marks = int(input("Enter marks: "))

#apply the condition of grade
if marks >= 90:
    print("A")
elif marks >= 75:
    print("B")
elif marks >= 50:
    print("C")
else:
    print("Fail")


```
