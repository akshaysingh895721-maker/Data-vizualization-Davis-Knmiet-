```python
def reverse_number(n):
    rev = 0
    while n > 0:
        digit = n % 10        # get last digit
        rev = rev * 10 + digit
        n = n // 10           # remove last digit
    return rev


# Example
num = int(input("Enter a number: "))
print("Reversed number:", reverse_number(num))
```
