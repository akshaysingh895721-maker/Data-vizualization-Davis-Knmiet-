```python
# to find the greatest digit present in given number
# Taking the number from the user
num = int(input("Enter a number: "))

max_digit = 0

while num > 0:
  #to find the last digit of the number
    digit = num % 10
    if digit > max_digit:
        max_digit = digit
        #Remove the last digit
    num = num // 10

print("Greatest digit:", max_digit)

```
