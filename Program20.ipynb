```python
#program to calculate simple interest

#taking input from the user
p = float(input("Enter P: "))
r = float(input("Enter R: "))
t = float(input("Enter T: "))

si = (lambda p, r, t: (p*r*t)/100)(p, r, t)
print("Simple Interest:", si)



```
